```python
data = data.drop(columns='education')
```

```python
data.head()
```

```python
freq_native_country = data['native-country'].value_counts()
```

```python
less_freq_countries = freq_native_country[freq_native_country<100]
```

```python
data['native-country'] = data['native-country'].replace(less_freq_countries.index, 'Other')
```

```python
data = pd.get_dummies(data, columns=['workclass', 'marital-status', 'occupation', 'relationship', 'gender', 'native-country', 'race', 'income'])
```

```python
data = data.drop(columns='income_<=50K')
```
